```python
def fibnum_rec(n):

    if n == 0 or n == 1:
       
       return n
    
    else:
        
        return fibnum_rec (n-2) + fibnum_rec (n-1)
    
print (fibnum_rec(10))


def fibnum_iter(n):

    a, b = 0, 1
    
    for i in range(n-1):

         result = a + b
         
         a, b = b, result

    return result 

print (fibnum_iter(10))


def fibnum_dynam(n, known = None):

    if known == None:
        
        known = {}

    if n == 0 or n == 1:
        
        return n

    if n in known:
      
        return known[n]

    else:
        
        result = fibnum_dynam(n-1, known) + fibnum_dynam(n-2, known)
         
        if n not in known:
            
            known[n] = result

    return known[n]
   
print (fibnum_dynam(10))
```
