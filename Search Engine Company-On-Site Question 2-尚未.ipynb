# Mine-probability isn't uniform

```python
import random

def dice5():
    
    return random.randint(1,5)

def dice7_simulation():
    
    a = []
    
    for i in range(1000):
        
        result = dice5() + dice5()
    
        while result > 7:
        
            result = dice5() + dice5()
        
        a.append(result)
    
    return a
    
```

```python
print (dice7_simulation())
```

# Instructor

```python
from random import randint
 
def dice5():
    return randint(1, 5)

def convert5to7():
    
    # Convert the combination to the range 1 to 25
    
    num = ( (dice5()-1) * 5 ) +  ( dice5() )      # -1是為了從0開始   &   *5是為了拉出五的間距 (因為dice5是從1到5)
                                                  #如此一來num才會從1到25
    if num > 21:
        
        # re-roll if we are out of range
        
        num = ( (dice5()-1) * 5 ) +  ( dice5() )

    return num %7 + 1                            # 1%7 ~ 21%7 => probability is uniform 
                                                 # +1是因為remainder有0
```

```python
convert5to7()
```
